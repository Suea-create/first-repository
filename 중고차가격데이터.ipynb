```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder
cars = pd.read_csv('cars.csv')
brands = pd.read_csv('brand.csv')

cars.columns = cars.columns.str.strip()
cars['Brand'] = cars['title'].str.split().str[0].str.upper()

brands = brands.rename({'title': 'brand_name'}, axis=1)
brands.columns = brands.columns.str.strip()
brands['brand_name'] = brands['brand_name'].str.upper()

merged_df = pd.merge(cars, brands, how='left', left_on='Brand', right_on='brand_name')
merged_df = merged_df.drop(['brand_name'],axis=1)
merged_df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
merged_df['Engine'] = le.fit_transform(merged_df['Engine'].astype(str))
merged_df['Emission Class'] = le.fit_transform(merged_df['Emission Class'].astype(str))
print(merged_df.head())
print(merged_df['Engine'].dtype)
print(merged_df['Emission Class'].dtype)
```

```python
categorical_cols = merged_df.select_dtypes(include=['object', 'category']).columns.tolist()
merged_df_encoded = pd.get_dummies(merged_df, columns=categorical_cols)
merged_df_encoded = merged_df_encoded.apply(pd.to_numeric, errors='coerce')
print(merged_df_encoded.head())
print(merged_df_encoded.dtypes.head(10))  
```

```python
merged_df_encoded[merged_df_encoded.select_dtypes('bool').columns] = merged_df_encoded.select_dtypes('bool').astype(int)
print(merged_df_encoded.head())
merged_df_encoded.info()
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
new_df = merged_df_encoded.select_dtypes(include=['int64', 'float64']).columns
scaled_data = ss.fit_transform(merged_df_encoded[new_df])
scaled_df = pd.DataFrame(scaled_data, columns= new_df)
print(scaled_df.head())
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np

ss = StandardScaler()
new_df = merged_df_encoded.select_dtypes(include=['int64', 'float64']).columns
scaled_data = ss.fit_transform(merged_df_encoded[new_df])
scaled_df = pd.DataFrame(scaled_data, columns= new_df)

from sklearn.decomposition import PCA
import pandas as pd

pca = PCA(n_components=0.7)
scaled_df_filled = scaled_df.dropna()
pca_result = pca.fit_transform(scaled_df_filled)

pca_df = pd.DataFrame(pca_result, columns=[f'PC{i+1}' for i in range(pca.n_components_)])
print(pca_df.head())
```
