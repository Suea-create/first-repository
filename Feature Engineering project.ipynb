```python
df=pd.read_csv('fraud.csv')
df.columns
```

```python
df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'], errors ='coerce')
df['dob'] = pd.to_datetime(df['dob'], errors='coerce')
```

```python
df.head()
```

```python
df['log_amt'] = np.log1p(df['amt']).clip(lower=0)
```

```python
df.columns
```

```python
df['hour'] = df['trans_date_trans_time'].dt.hour
df['dayofweek'] = df['trans_date_trans_time'].dt.dayofweek
df['is_weekend'] = df['dayofweek'].isin([5,6]).astype(int)
df['is_night'] = df['hour'].between(0,6).astype(int)
```

```python
age_days =  (df['trans_date_trans_time']- df['dob']).dt.days
```

```python
df['age'] = (age_days/ 365.25).astype(float)
```

```python
drop_cols = ['trans_date_trans_time', 'cc_num', 'amt', 'first', 'last', 'street', 'city', 'state', 'zip','dob', 'trans_num', 'unix_time']
cc_df = df.drop(columns = [c for c in drop_cols if c in df.columns] , errors='ignore')
cc_df.head()
```

```python
#isna()sum()으로 결측치를 찾아보았으나 결측치가 없었다. 그러나 혹시 모를 에러를 방지하기위해 결측치를 숫자와 문자열로 나누어 처리한다.
num_cols = cc_df.select_dtypes(include = [np.number]).columns.tolist()
cat_cols = cc_df.select_dtypes(include = ['object', 'category']).columns.tolist()
for c in num_cols:
    if c != 'is_fraud':
        cc_df[c] = cc_df[c].fillna(cc_df[c].median())
for c in cat_cols:
    cc_df[c] = cc_df[c].fillna('unknown')
cc_df.head()
```

```python
one_hot_cols = [col for col in ['gender'] if col in cc_df.columns]
cc_df = pd.get_dummies(cc_df, columns= one_hot_cols, drop_first=True)
cc_df.head()
```
